Great job! You just learned how to tune the 2 main hyperparameters responsible for initialising the k-means clusters. You have seen in this exercise that increasing the number of iterations (n_init) didn’t have much impact on the clustering result for this dataset. In this case, it is better to use a lower value for this hyperparameter as it will speed up the training time. But for a different dataset you may face the case where the results differ drastically depending on the n_init value. In this case, you will have to find a value of n_init that is not too small but not too big. You want to find the sweet spot where the results are not changing much compared to the last result obtained with a different value.
